## Conversión de diccionario a dataframe
Diccionarios
  1. no tienen indexación
  2. Poseen clave única y un valor asociado a la clave
  3. son mutables

````python
diccionario={"enero":22,"febrero":15,"marzo":70}
type(diccionario)
````

````python
ventas1=pd.Series(diccionario)
ventas1
````

````python
alex=pd.Series(diccionario, index=["enero","febrero","marzo","abril"])
alex
````

````python
# Diccionario de elementos
elementos = {
    "número atómico": [1, 6, 47, 88],
    "Masa atómica": [1, 12, 107, 226],
    "Familia": ["No metal", "No metal", "Metal", "Metal"]
}

df_elementos = pd.DataFrame(elementos,index=["Hidrógeno", "Carbono", "Plata", "Radón"])
df_elementos.index.name = "Elementos"

df_elementos
````

````python
# Diccionarios de medidas
medida_2020 = {"Ag": 2, "Au": 5, "Cu": 3, "Pt": 2}
medida_2021 = {"Ag": 4, "Au": 3, "Cu": 6, "Pt": 1}
medida_2022 = {"Ag": 8, "Au": 2, "Cu": 15, "Pt": 0}
medida_2023 = {"Ag": 5, "Au": 1, "Cu": 0, "Pt": 5}
medida_2024 = {"Ag": 7, "Au": 0, "Cu": 3, "Pt": 2}
medida_2025 = {"Ag": 9, "Au": 1, "Cu": 3, "Pt": 5}

# Crear un DataFrame a partir de las medidas y trasponemos para que los años sean filas
df_medidas = pd.DataFrame({
    "2020": medida_2020,
    "2021": medida_2021,
    "2022": medida_2022,
    "2023": medida_2023,
    "2024": medida_2024,
    "2025": medida_2025
}).T

df_medidas. index.name="Año"
df_medidas
````
